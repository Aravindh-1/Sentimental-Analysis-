Importing Transformers and Sentimental analysis library to differentiate the reviews as Positive and Negative.


```python
import transformers
from transformers import pipeline
```

```python
sentimental_analysis=pipeline('sentiment-analysis')
rating_value=[]
for i in (analysis_data['review']):
  rating_value.append(sentimental_analysis(i))
print(rating_value)
analysis_data['rating_score']=rating_value
```

```python
analysis_data.drop(analysis_data.columns[analysis_data.columns.str.contains('Unnamed', case=False)], axis=1, inplace=True)
```

```python
positive_labels = []
for sublist in rating_value:
    for entry in sublist:
        if isinstance(entry, dict) and entry.get('label') == 'POSITIVE':
            positive_labels.append(entry['label'])
pos=len(positive_labels)
print(positive_labels)
```

```python
negative_labels = []
for sublist in rating_value:
    for entry in sublist:
        if isinstance(entry, dict) and entry.get('label') == 'NEGATIVE':
            negative_labels.append(entry['label'])
neg=len(negative_labels)
print(negative_labels)
```

There are 24 Positive and 26 Negative reviews in the dataset, now plotting the numbers for visualization



```python
total=analysis_data['rating_score'].count()
print(total)
```

Generating Pie Chart with Positive and Negative Ratings

```python
percent_neg= (neg/total)*100
percent_pos= (pos/total)*100
data = [percent_pos,percent_neg]
labels = ["Positive_Percentage", "Negative_Percentage"]
plt.pie(data, labels=labels, autopct="%1.2f%%")
plt.axis("equal")
plt.title("Percentage of ratings")
plt.show()


```
